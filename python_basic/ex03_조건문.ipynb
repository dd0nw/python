#### if else 실습
- if문에 있는 조건식에 해당하지 않으면 else문을 실행
- else문에는 조건식이 필요하지 않음

```python
money = 11000
if money >= 10000:
    print('택시를 탄다~ 씽씽')

else:
    print('버스를 탄고 간다 붕붕 ~')
```

#### if else 실습 예제
1. 키보드로 변수 num을 입력받고 num이 3의 배수이면서 5의 배수이면!"3과5의배수입니다."를 출력하고, 그게 아니라면 "3과 5의 배수가 아닙니다."를 출력하시오.

```python
num=int(input('정수 입력 >> '))

if num%3 == 0 and num%5 == 0:
    print(f"{num}은 3과 5의 배수입니다.")

else:
    print(f"{num} 3과 5의 배수가 아닙니다.")
```

2. 대박이는 마스크를 대량 구매했습니다. 주변 친구들에게 소분해서 나눠주기 위해 포장지를 사려 합니다. 포장지는 마스크를 각각 8개, 5개식 넣을 수 있는 8개들이, 5개들이 2종류가 있습니다. 8개들이는 포장지가 비싸서 반드시 꽉곽 채워 포장할 때만 사용하고, 만약 마스크가 남는다면 5개들이에 포장하려고 합니다. 과연 포장지는 각각 몇개가 필요할까요?
(단, 마스크는 남는 마스크 없이 전부 포장해야 합니다.)

```python
a=int(input('마스크 개수 입력>> '))

b=a//8 #8개봉지 - 초기화
c=int(a%8/5) #5개봉지
    
if a%8 <= 5:
    c=1
if a%8==0:
    c=0
if a%8 >5:
    c=2

print(f'필요한 마스크 8개들이 개수 : {b}, 5개들이 개수 : {c}')
```

#### if elif else 실습예제
1. 변수 num1과 num2에 숫자를 입력 받아서 크기를 비교하시오.

```python
num1 = int(input('정수입력1 >> '))
num2 = int(input('정수입력2 >> '))

if num1 < num2:
    print('첫 번째 정수가 더 작습니다.')
elif num1 < num2:
    print('두 번째 정수가 더 작습니다.')
else:
    print('두 정수가 같습니다')
```

2. 변수 score에 점수를 입력 받아서 학점 관리 프로그램을 만들어보자.
 - 90점 이상인 점수에는 'A학점', 80점 이상인 점수에는 'B학점', 70점 이상인 점수에는 'C학점', 60점 이상인 점수에는 'D학점', 60점 미만인 점수는 'F학점'

```python
score = int(input('점수 입력 >> '))

if score >= 90:
    print(f'{score}점은 A학점 입니다.')
elif score >= 80:
    print(f'{score}점은 B학점 입니다.')
elif score >= 70:
    print(f'{score}점은 C학점 입니다.')
elif score >= 60:
    print(f'{score}점은 D학점 입니다.')
else:
    print(f'{score}점은 F학점 입니다.')
```
