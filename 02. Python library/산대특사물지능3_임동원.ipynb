```python
#4
score_list=[['김미희', 89,78,99],
           ['서대희',67,88,77],
           ['강예진',100,55,96]]
print('='*50)
print('이름    Python    Java    DB    평균    학점')
print('='*50)


for i in range(3):
    avg1=int((score_list[i][1]+score_list[i][2]+score_list[i][3])/3)
    score_list[i].append(avg1)
    if avg1>= 90:
        grade='A'
    elif avg1>= 80:
        grade='B'
    elif avg1>= 70:
        grade='C'
    elif avg1>= 60:
        grade='D'
    else:
        grade='F'
    score_list[i].append(grade)
        
        
for k in range(3):
    print(score_list[k][0],' ',score_list[k][1],'     ', score_list[k][2], '    ',score_list[k][3],'    ',score_list[k][4], '   ',score_list[k][5])
    
print('='*50)

lst=[]
    
for j in range(1,4):
    avg2=int((score_list[0][j]+score_list[1][j]+score_list[2][j])/3)
    lst.append(avg2)
    
print('평균', '   ',lst[0], '       ', lst[1],'      ', lst[2])
```

```python
#5
import matplotlib.pyplot as plt

X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]

plt.plot(X,Y1, color='red', label='Y1')
plt.plot(X,Y2, color='blue', label='Y2')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
plt.legend()

plt.show()
```

```python
# 6번

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')

temp= data[(data['사고유형_대분류'] == '차대사람') & (data['주야'] == '야간')]
temp2=temp['요일'].value_counts()
temp2.sort_index()
y = temp2[['월','화','수','목','금','토','일']]

from matplotlib import rc
rc('font', family='Malgun Gothic')

x=range(7)
plt.bar(y.index, y)

plt.ylim(0,200)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')

plt.show()
```
