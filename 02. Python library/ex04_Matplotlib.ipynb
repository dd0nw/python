## matplotlib
 - 2D 그래프로 시각화가 가능한 라이브러리
 - 파이썬에서 데이터를 차트나 플롯으로 그려줌
 - 차트의 종류는 산점도, 바, 파이, 히스토그램, 꺾은선 그래프 등 다양하게 존재

### 데이터 시각화란?
 - 광범위하게 분산된 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트 등으로 정리하는 것
 - 시각화를 통해 데이터의 특징을 쉽게 파악할 수 있다.

```python
# matplotlib 패키지 불러온 후에 pyplot이라는 라이브러리를 사용하겠다.
# 이름은 plt라고 불러주겠다
import matplotlib.pyplot as plt
```

### 기본적인 그래프 그리기

```python
# Line Plot (선 그래프)
y = [2,12,6]
plt.plot(y)

# plot(좌표의 y요소)
# x좌표는 0부터 1씩 증가하며 자동으로 할당

# plt.show() --> 그동안 그려왔던 plot들을 출력하겠다 !
# print()와 같은 기능을 한다 --> 그래프를 그릴 때는 항상 마지막에 show로 마무리 !!
plt.show()
```

```python
# x축, y축 전부 설정해서 그려보자
# 선 스타일도 입혀보자
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y, ls='-.')
plt.show()
```

```python
# x축, y축 설정해서 그리고 marker 찍기
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y, ls='--', marker='o')
plt.show()
```

```python
# x축, y축 설정해서 선 두께 설정
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y, ls='--', marker='o', lw = 5)
plt.show()
```

```python
# x축, y축 설정해서 선 색깔 설정
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y, ls='--', marker='o', lw= 5, color='r')
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

# 속성값
# 1. 라인의 스타일 설정 : ls(Line Style) -> 문자열
# 2. 라인의 두께 설정 : lw(Line Width) -> 숫자
# 3. 선의 색 바꾸기 : c(Color) ->  선색과 마커의 색까지 모두 변경
# 4. 마커 내부 색 바꾸기 : mfc(marker face color) -> 문자열
# 5. 마커 선 색깔 : mec(marker edge color) -> 문자열
# 6. 마커의 크기 : ms -> 숫자
# 7. 마커의 선 굵기 : mew(marker edge width) -> 숫자
# 8. 

plt.plot(x,y, ls='--', marker='o', lw= 5, color='r', mfc='m')
plt.show()

# RGB = Red, Green, Blue
```

### 선 그래프 실습
 - x = np.arange(8)
 - y = [5,2,3,6,8,7,5,7]

```python
import numpy as np

x=np.arange(8)
y=[5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='--',c='g',marker='o',mec='g')
plt.show()
```

```python
# $달러 기호 안에 넣어주고 싶은 기호를 사용하면 마커로 사용 가능 !
plt.plot(x,y,ls='-.',c='y',marker='$☆$',mec='skyblue', lw=5, ms=20, mfc='w', mew=3)
plt.show()
```

```python
# $달러 기호 안에 넣어주고 싶은 기호를 사용하면 마커로 사용 가능 !
plt.plot(x,y,ls='-.',c='#FFC596',marker='$☆$',mec='skyblue', lw=5, ms=20, mfc='w', mew=3)

# color -> 색상 코드를 직접 입력해도 무방하다(문자열 형태)
# 그래프의 숫자 표시 범위 지정
# 범위를 지정하여 그래프를 축소하거나 확대할 수 있다.

plt.xlim(-2, 10)
plt.ylim(-5, 12)

plt.show()
```

```python
plt.plot(x,y,ls='-.',c='#FFC596',marker='$☆$',mec='skyblue', lw=5, ms=20, mfc='w', mew=3)

# 그리드 그리기 (그리드 : 모눈종이, 격자눈금) -> 그래프를 보기 쉽게 하기 위해
plt.grid()

# 눈금 출력방법 -> ticks
plt.yticks(y) #y데이터를 기준으로 눈금을 그려달라

plt.show()
```

### 여러개 차트를 하나의 공간에 그리기

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')

# 범례 설정하기
# 범례 : (참고) 데이터의 종류를 표시하기 위한 텍스트
plt.legend()

# 차트 제목 설정하기
# loc : 위치설정 (left, center, right)
plt.title('차트', loc='right')

# x축과 y축에 이름도 집어넣어보자
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# 한글 인코딩 하는 방법 - matplotlib 한글이 깨졌을 때 사용!
from matplotlib import rc
rc('font', family='Malgun Gothic')
# font : 폰트를 설정하겠다 !
# family : 글꼴 집합이라는 의미로 글꼴의 이름을 적어주면 된다.
```

```python
# bar 그래프
x = [0,1,2,3,4,5]
y = [80,95,70,60,50,90]

plt.bar(x,y)
plt.xlabel('강의장')
plt.ylabel('점수')

plt.show()
```

```python
#barh 그래프

x = [0,1,2,3,4,5]
y = [80,95,70,60,50,90]

plt.barh(x,y)
plt.ylabel('강의장')
plt.xlabel('점수')

plt.show()
```

```python
# bar 그래프

x = [0,1,2,3,4,5]
y = [80,95,70,60,50,90]

plt.bar(x,y)

plt.xticks(x, ['1강의장', '2강의장', '3강의장', '4강의장','5강의장', '6강의장'])
plt.yticks(y)

# ylim를 설정해서 분포된 데이터 확대
plt.ylim(45,100)

plt.xlabel('강의장')
plt.ylabel('정수')

plt.title('CGI 강의장별 시험성적', loc='right')

plt.show()
```

### 공공데이터 분석 및 시각화
 - 2017년 전국 교통사고 데이터

```python
# 데이터 분석에 필요한 라이브러리 로딩

# pandas, numpy, matplotlib.pyplot

import numpy as np # 수치해석 전용 라이브러리
import pandas as pd # 데이터를 표 형태로 제공해주는 라이브러리
import matplotlib.pyplot as plt # 데이터 시각화를 도와주는 라이브러리
```

```python
# 데이터 로딩
data=pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')
```

```python
# 데이터 규모(정보) 확인
# info() : 데이터프레임에서 컬럼이름, NaN(NULL)값이 아닌 갯수, 데이터 타입
data.info()
```

```python
# 데이터 크기 확인
data.shape
```

```python
# 데이터 프레임 최상단 5개만 출력하는 방법
# 기본값 5개
# 원하는 만큼 숫자를 입력해서 출력을 조절
data.head(3)
```

```python
# 데이터 프레임 최하단 5개만 출력하는 방법
data.tail()
```

```python
# 0건 : 몇개 ?
# 1건 : 몇개?
# 2건 : 몇개?
# value_coumts() : 하나의 컬럼안에 있는 유니크 값들과 그 값들의 개수를 각각 출력
data['부상신고자수'].value_counts()
```

### 요일별 사고건수 카운트
 - 요일 순서대로 정렬 후 변수에 저장(y)

```python
# 요일별 발생한 사고건수를 카운트 해보자
temp=data['요일'].value_counts()
temp
```

```python
# 월~금요일까지 순서대로 정렬
temp.sort_index() # 정렬을 하더라도 가나다순으로 정렬이 되어버린다.
```

```python
# 유니크 값으로 설정 후 변수에 담아주기
y = temp[['월','화','수','목','금','토','일']]
y
```

### 시각화
- 바 그래프

```python
# y축 = y -> 범위 500 ~620
# x축 = 7
# 그래프 제목 : 요일별 사망교통사고 건수
# x축 : 요일
# y축 : 사고건수

plt.bar(y.index, y)
plt.yticks(y)


plt.ylim(500,620)
plt.title('요일별 사망교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고건수')

plt.show()
```

```python
x=range(7)
plt.bar(x,y)
plt.xticks(x,['월','화','수','목','금','토','일'])
plt.yticks(y) # y축 값이 정확하게 어떤 값을 가지는지 직관적으로 나타내기 위함

plt.ylim(500,620) # y축 범위 : 500 ~ 620

plt.xlabel('요일')
plt.ylabel('사고건수                  ', rotation=0) # 사고건수라는 글자를 90도 세워보자
#rotation : 회전이라는 뜻으로 y라벨에 해당하는 글자를 얼마나 회전할 것인지 기입(숫자)
plt.title('요일별 사망 교통사고 건수', loc='right')

plt.show()
```
